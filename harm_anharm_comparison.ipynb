```python
#libraries and packages needed:
import pandas as pd
import numpy as np
import ast
import os
```

```python
# levels of theory:
lots = ['B971', 'B3LYP', 'wB97XD', 'M062X', 'PBE1PBE', 'TPSSh', 'B2PLYP', 'B2PLYPD']

#dict. to store frequency data:
freq_data = {'Model Chemistry' : [],
            'Model Chemistry Class' : [],
            'Method' : [],
            'Basis Set' : [],
            'Molecule' : [],
            'Experimental Frequnecy' : [],
            'Harmonic Frequnecy' : [],
            'Anharmonic Frequnecy' : [],
            'Harmonic Intensity' : [],
            'Anharmonic Intensity' : []
            }

freq_data = pd.DataFrame(freq_data) #from dict. to df

#dict. to store model chemistry specific scaling factors data:
sf_data = {'Model Chemistry' : [],
           'Low Harmonic SF' : [],
           'Mid Harmonic SF' : [],
           'High Harmonic SF' : [],
           'Low Anharmonic SF' : [],
           'Mid Anharmonic SF' : [],
           'High Anharmonic SF' : []
          }

sf_data = pd.DataFrame(sf_data) #from dict. to df

#dict. to store model chemistry frequency comparison data:
harm_anharm_comparison = {'Model Chemistry' : [],
                          'Model Chemistry Class' : [],
                          'Method' : [],
                          'Basis' : [],
                          'Molecule' : [],
                          'Experimental Frequnecy' : [],
                          'Harmonic Frequnecy' : [],
                          'Scaled Harmonic Frequnecy' : [],
                          'Anharmonic Frequnecy' : [],
                          'Scaled Anharmonic Frequnecy' : [],
                          'Harmonic Intensity' : [],
                          'Anharmonic Intensity' : []
                         }

harm_anharm_comparison = pd.DataFrame(harm_anharm_comparison) #from dict. to df

#directories:
general_directory = '\\Users\\natia\\OneDrive\\Pulpit\\benchmark paper'
harm_freq_files = 'calculations\\final_ordered_by_lot\\harmonic'
anharm_freq_files = 'calculations\\final_ordered_by_lot\\anharmonic'
```

## frequency data:

```python
#fill the mc, mc class, method, basis set, molecule, exp freq, harm freq, harm int
for lot in lots : 
    
    harm_freq_file = pd.read_csv(f'{general_directory}\\{harm_freq_files}\\{lot}.csv')
    anharm_freq_file = pd.read_csv(f'{general_directory}\\{anharm_freq_files}\\{lot}.csv')
    
    for mol, lot, bs, expfreq, harmfreq, anharmfreq, harmint, anharmint in zip(harm_freq_file['Formula'], harm_freq_file['Method'], harm_freq_file['Basis'], harm_freq_file['ExpFreq'], harm_freq_file['CalcFreq'], anharm_freq_file['CalcFreq'], harm_freq_file['Intensity'], anharm_freq_file['Intensity']) :
        
        mc = lot + '/' + bs
        
        if lot == 'B2PLYP' or lot == 'B2PLYPD' :
            lot_class = 'DH'
        else :
            lot_class = 'hybrid'
            
        if bs == 'pc-2' or bs == 'pcseg-2' or bs == 'AUG-pcseg-2' or bs == 'Def2TZVPD' or bs == 'cc-pVTZ' :
            bs_class = 'TZ'
        else : 
            bs_class = 'DZ'
        
        mc_class = lot_class + '/' + bs_class
        
        freq_data.loc[len(freq_data.index)] = [mc, mc_class, lot, bs, mol, expfreq, harmfreq, anharmfreq, harmint, anharmint] 
```

## scaling factors data:

```python
#fill the sf data:
harm_sf_file = pd.read_csv(f'{general_directory}\\results\\TOTAL_frequency_specific_harmonic.csv')
anharm_sf_file = pd.read_csv(f'{general_directory}\\results\\TOTAL_frequency_specific_anharmonic.csv')

for mc, lh_sf, mh_sf, hh_sf, la_sf, ma_sf, ha_sf in zip(harm_sf_file['ModChem'], harm_sf_file['SF_Low_mean'], harm_sf_file['SF_Mid_mean'], harm_sf_file['SF_High_mean'], anharm_sf_file['SF_Low_mean'], anharm_sf_file['SF_Mid_mean'], anharm_sf_file['SF_High_mean']) :

    sf_data.loc[len(sf_data.index)] = [mc, lh_sf, mh_sf, hh_sf, la_sf, ma_sf, ha_sf] 
```

## comparison data:

= scaling frequencies

```python
for mc_all, mc_class, lot, bs, mol, expfreq, harmfreq, anharmfreq, harmint, anharmint in zip(freq_data['Model Chemistry'], freq_data['Model Chemistry Class'], freq_data['Method'], freq_data['Basis Set'], freq_data['Molecule'], freq_data['Experimental Frequnecy'], freq_data['Harmonic Frequnecy'], freq_data['Anharmonic Frequnecy'], freq_data['Harmonic Intensity'], freq_data['Anharmonic Intensity']) :
    
    for mc_sf, lh_sf, mh_sf, hh_sf, la_sf, ma_sf, ha_sf in zip(harm_sf_file['ModChem'], harm_sf_file['SF_Low_mean'], harm_sf_file['SF_Mid_mean'], harm_sf_file['SF_High_mean'], anharm_sf_file['SF_Low_mean'], anharm_sf_file['SF_Mid_mean'], anharm_sf_file['SF_High_mean']) :
        
        if mc_all == mc_sf :
            
            if expfreq < 1000 :
                
                s_harmfreq = harmfreq * lh_sf
                s_anharmfreq = anharmfreq * la_sf                
            
            elif expfreq >= 1000 and expfreq <= 2000 :
                
                s_harmfreq = harmfreq * mh_sf
                s_anharmfreq = anharmfreq * ma_sf                
            
            elif expfreq > 2000 :
                
                s_harmfreq = harmfreq * hh_sf
                s_anharmfreq = anharmfreq * ha_sf
                
            else :
                
                print(mc_all + expfreq)
            
            harm_anharm_comparison.loc[len(harm_anharm_comparison.index)] = [mc_sf, mc_class, lot, bs, mol, expfreq, harmfreq, s_harmfreq, anharmfreq, s_anharmfreq, harmint, anharmint]
```
