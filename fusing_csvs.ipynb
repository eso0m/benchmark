# Fusing run 1 & 2:

```python
lots = ['B971', 'B3LYP', 'wB97XD']
bss = ['6-31Gs', '6-31Gss', 'pc-1', 'pcseg-1', 'AUG-pcseg-1', 'pc-2', 'pcseg-2', 'AUG-pcseg-2', 'Def2SVP', 'Def2SVPD', 'Def2TZVPD', 'cc-pVDZ', 'AUG-cc-pVDZ', 'cc-pVTZ']

for lot in lots :
    
    for bs in bss :

        First = pd.read_csv(f'\\Users\\natia\\OneDrive\\Pulpit\\benchmark paper\\calculations\\output\\run1\\{lot}_{bs}.csv')
        
        Second = pd.read_csv(f'\\Users\\natia\\OneDrive\\Pulpit\\benchmark paper\\calculations\\output\\run2\\{lot}_{bs}.csv')
        count = 0 

        for i in Second.index :
    
            Second['Unnamed: 0'][i] = len(First) + count
            count += 1
    
        First = pd.concat([First,Second])
        
        First.to_csv(f'\\Users\\natia\\OneDrive\\Pulpit\\benchmark paper\\calculations\\final\\{lot}_{bs}.csv',index=None)
```

# Fusing hybrids and double hybrids:

```python
initial_path = '\\Users\\natia\\OneDrive\\Pulpit\\benchmark paper\\results'

resulttype = ['frequency_specific', 'global']
calctype = ['harmonic', 'anharmonic']

count = 0 

for a in resulttype :
    
    for b in calctype : 
        
        first = pd.read_csv(f'{initial_path}\\{a}_{b}_hybrids.csv')
        second = pd.read_csv(f'{initial_path}\\{a}_{b}_doublehybrids.csv')
        
        for i in second.index :

            second['MC_order'][i] = len(first) + count
            count = count + 1

        first = pd.concat([first, second])
        first.to_csv(f'{initial_path}\\TOTAL_{a}_{b}.csv', index = None)
```
